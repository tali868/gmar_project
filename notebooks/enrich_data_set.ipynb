## enrich size and name

```python
body = {"subClasses": ["Container Vessel"]}
all_container_vessels_response = requests.post(
    'http://k8s-prod-java-vessels.windward.com/api/v1/vessels/getAllVesselsBy', json=body)

all_container_vessels = json.loads(all_container_vessels_response.text)
```

```python
clean_vessels = []
for vessel in all_container_vessels:
    new_vessel = {"vessel_id": vessel["id"], "name": vessel["name"]}
    new_vessel["size"] = vessel.get("size")
    new_vessel["flag"] = vessel.get("flag")
    clean_vessels.append(new_vessel)
vessels_df = pd.DataFrame(clean_vessels)
```

```python
len(clean_vessels)
```

```python
reported_pc_size_df = pd.merge(reported_w_pc_df, vessels_df, how="left", on=["vessel_id"])
if len(reported_pc_size_df["vessel_id"].unique()) != len(reported_pc_size_df["vessel_id"].unique()):
    raise Exception("MERGING ERROR")
else:
    print("Unique vessel cound: {}".format(len(reported_w_pc_df["vessel_id"].unique())))
```

```python
# ---------------------------- #
```

```python
def get_polygon_to_name(polygon_id):
    host = mu.MONGO_PRODUCTION2_HOST
    db = "atlas"
    collection = "areas"
    data = mu.read_from_mongo(host, db, collection, query={"_id": ObjectId(polygon_id)})
    return {"polygon_id": polygon_id, "port_name": data[polygon_id]["properties"].get("title"), "port_country": data[polygon_id]["properties"].get("country", None)} if data else {"polygon_id": polygon_id, "port_name": None, "port_country": None}
    
```

```python
polygon_names = [get_polygon_to_name(polygon_id) for polygon_id in list(
    reported_pc_size_df["next_pc"].unique()) if isinstance(polygon_id, str)]
```

```python
left_polygon_ids = set(reported_pc_size_df["resolved_dest"].unique()) - set(reported_pc_size_df["next_pc"].unique())
left_polygon_names = [get_polygon_to_name(polygon_id) for polygon_id in left_polygon_ids if isinstance(polygon_id, str)]
```

```python
polygon_names_df = pd.DataFrame(polygon_names + left_polygon_names)
```

```python
polygon_names_df
```

```python
reported_w_next_pc_df = pd.merge(reported_pc_size_df, polygon_names_df, how="left", left_on=["next_pc"], right_on=["polygon_id"])
reported_w_next_pc_df = reported_w_next_pc_df.drop(columns=["polygon_id"]).rename(columns={"port_name": "next_pc_name", "port_country": "next_pc_country"})
```

```python
reported_full_pc_df = pd.merge(reported_w_next_pc_df, polygon_names_df, how="left", left_on=["resolved_dest"], right_on=["polygon_id"])
reported_full_pc_df = reported_full_pc_df.drop(columns=["polygon_id"]).rename(columns={"port_name": "resolved_name", "port_country": "resolved_country"})
```
